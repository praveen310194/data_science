```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import copy
```

```python
df = pd.read_csv("D:/practice/train_practice.csv")
df
```

```python
df_row = pd.concat([df, df*2, df*3, df*4, df*5])
```

```python
df_row
```

```python
df_row.head()
```

```python
X=df_row.drop('total',axis=1)
Y=df_row['total']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train, x_cv, y_train, y_cv = train_test_split(X,Y, test_size =0.3,random_state=8)
```

```python
model = LinearRegression() 
model.fit(x_train, y_train)
result = model.score(x_cv, y_cv)
print("Accuracy: %.3f%%" % (result*100.0))
```
